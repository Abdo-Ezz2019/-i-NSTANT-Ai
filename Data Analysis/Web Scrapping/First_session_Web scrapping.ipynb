```python
from bs4 import BeautifulSoup 
import requests
```

```python
#   API     
url= "https://www.yallakora.com/match-center?date=8/19/2025#days"
response = requests.get(url)
response




```

```python
html_content = response.content # HTML content
html_content

```

```python
soup = BeautifulSoup(html_content, 'html.parser') # Parse HTML content
soup
```

```python
championship = soup.find_all('div',{"class":"matchCard"})
championship
```

```python
len(championship) # Number of matches found
```

```python
championship[0]
```

```python
champ_Name=[]
for i in range(len(championship)):
    champ_Name.append(championship[i].find("h2").text.strip()) # Match title
print(champ_Name)
```

```python
for i in range(len(championship)):
    champ_Name=championship[i].find("h2").text.strip() # Match title
    print(champ_Name)
```

```python
matches=[]
for i in range(len(championship)):
    matches.extend(championship[i].find_all('div',{"class":"teamsData"}))
matches
```

```python
len(matches) # Number of teams in the match
```

```python
teams=[]
scores=[]
times=[]

for i in range(len(matches)):
    teamA=matches[i].find("div",{"class":"teamA"}).text.strip()
    teamB=matches[i].find("div",{"class":"teamB"}).text.strip()
    teams.append((teamA, teamB))
    score=matches[i].find_all("span",{"class":"score"})
    time=matches[i].find('span',{"class":"time"}).text.strip()
    scores.append((score[0].text, score[1].text))
    times.append(time)

print("Match Details:")
for i in range(len(matches)):
    print(f"Match {i + 1}: {teams[i][0]} vs {teams[i][1]}")
    print(f"Score: {scores[i][0]} - {scores[i][1]}")
    print(f"Time: {times[i]}")

```
